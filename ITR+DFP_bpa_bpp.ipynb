```python
import pandas as pd
import requests
import zipfile
import io
import time
import matplotlib.pyplot as plt

x = int(input('Ano Inicial: '))
y = int(input('Ano Final + 1: '))

lista_planilha =[]

tipo_planilha = 'a'
while tipo_planilha != ' ':
    tipo_planilha = input('Planilha: BPA, BPP, DFC_MD, DFC_MI, DRA, DRE ou DVA ? ') 
    lista_planilha.append(tipo_planilha)
lista_planilha.remove(' ')
    
lista_itr_df = []
lista_dfp_df = []

for planilha in lista_planilha:
    for ano in range(x,y):
        link = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/itr_cia_aberta_{}.zip'.format(ano)
        arquivo = 'itr_cia_aberta_{}_con_{}.csv'.format(planilha, ano)
        print('Download ITR {} {}...'.format(planilha, ano))
        r = requests.get(link)
        time.sleep(0.5)
        print('Deszipando ITR {} {}...'.format(planilha, ano))
        zf = zipfile.ZipFile(io.BytesIO(r.content))
        time.sleep(0.5)
        zf = zf.open(arquivo)
        time.sleep(0.5)
        lines = zf.readlines()
        time.sleep(0.5)
        lines = [i.strip().decode('ISO-8859-1') for i in lines]
        time.sleep(0.5)
        lines = [i.split(';') for i in lines]
        time.sleep(0.5)
        df_itr = pd.DataFrame(lines[1:], columns=lines[0])
        
        lista_itr_df.append(df_itr)

for planilha in lista_planilha:
    for ano in range(x,y):
        link = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/dfp_cia_aberta_{}.zip'.format(ano)
        arquivo = 'dfp_cia_aberta_{}_con_{}.csv'.format(planilha, ano)
        print('Download DFP {} {}...'.format(planilha, ano))
        r = requests.get(link)
        time.sleep(0.5)
        print('Deszipando DFP {} {}...'.format(planilha, ano))
        zf = zipfile.ZipFile(io.BytesIO(r.content))
        time.sleep(0.5)
        zf = zf.open(arquivo)
        time.sleep(0.5)
        lines = zf.readlines()
        time.sleep(0.5)
        lines = [i.strip().decode('ISO-8859-1') for i in lines]
        time.sleep(0.5)
        lines = [i.split(';') for i in lines]
        time.sleep(0.5)
        df_dfp = pd.DataFrame(lines[1:], columns=lines[0])
        
        
        
        lista_dfp_df.append(df_dfp)
    
arquivo1 = pd.DataFrame()
arquivo2 = pd.DataFrame()
for data in lista_itr_df:
    arquivo1 = pd.concat([arquivo1, data])

for data1 in lista_dfp_df:
    arquivo2 = pd.concat([arquivo2, data1])
    
arquivo = pd.DataFrame()
arquivo = pd.concat([arquivo1, arquivo2])

arquivo = arquivo[arquivo['ORDEM_EXERC']=='ÚLTIMO']
arquivo = arquivo[['DENOM_CIA','DT_FIM_EXERC','DS_CONTA','VL_CONTA', 'GRUPO_DFP']]
arquivo['VL_CONTA'] = pd.to_numeric(arquivo['VL_CONTA'], errors='coerce')
pd.options.display.float_format = "{:,.2f}".format
arquivo = arquivo.sort_values(by= 'DT_FIM_EXERC')
arquivo.set_index('DT_FIM_EXERC', inplace=True)

display(arquivo)
```
